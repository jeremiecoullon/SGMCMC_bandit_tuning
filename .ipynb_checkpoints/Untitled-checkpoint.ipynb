```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import jax.numpy as jnp
from jax import random
from sgmcmcjax.samplers import build_sgld_sampler, build_sgldCV_sampler


# define model in JAX
def loglikelihood(theta, x):
    return -0.5*jnp.dot(x-theta, x-theta)

def logprior(theta):
    return -0.5*jnp.dot(theta, theta)*(1/10_000)

# generate dataset
N, D = 10_000, 1
key = random.PRNGKey(0)
X_data = random.normal(key, shape=(N, D))
```

```python
batch_size
```

```python

# build sampler
batch_size = N
dt = 1e-5
ula_sampler = build_sgld_sampler(dt, loglikelihood, logprior, (X_data,), batch_size)

# run sampler
Nsamples = 10_000
samples_ula = ula_sampler(key, Nsamples, jnp.zeros(D))

```

```python

# build sampler
batch_size = int(0.01*N)
print(batch_size)

dt = 1e-5
my_sampler = build_sgld_sampler(dt, loglikelihood, logprior, (X_data,), batch_size)
# my_sampler = build_sgldCV_sampler(dt, loglikelihood, logprior, (X_data,), batch_size, 0.0)

# run sampler
Nsamples = 10_000
samples = my_sampler(key, Nsamples, jnp.zeros(D))
```

```python
plt.plot(samples[100:,0], label="sgld")
plt.plot(samples_ula[100:,0], label="ula")

# plt.plot(samples[100:,0] - samples_ula[100:,0])

plt.legend()
plt.show()
```

```python
plt.hist(samples[100:,0], label="sgld", bins=20)
plt.hist(samples_ula[100:,0], label="ula", bins=20, alpha=0.7)
plt.legend()
plt.show()
```
